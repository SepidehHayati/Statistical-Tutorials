# Generating random data to display various data distributions

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Normal Distribution

```python
# Generates 100,000 random numbers from a normal distribution with mean 0 and standard deviation 1.
data_normal = stats.norm.rvs (size = 100000, loc = 0, scale = 1)
```

```python
type (data_normal)
```

```python
# Visualizes the distribution with a histogram.
sns.displot(data_normal, bins = 100, kde = False)
```

## 2. Uniform Distribution

```python
# Generates 100,000 random numbers from a uniform distribution ranging from 0 to 10.
data_uniform = stats.uniform.rvs (size = 100000, scale = 10)
```

```python
#Displays the distribution using a histogram.
sns.displot (data_uniform, bins = 100, kde = False)
```

## 3. Bernoulli Distribution

```python
# Generates 1,000 random numbers from a Bernoulli distribution with success probability p=0.2.
data_bernoulli = stats.bernoulli.rvs (size = 1000, p = 0.2)

#Prints the generated values (a series of 0s and 1s).
print (data_bernoulli)
```

```python
# Plots the frequency of 0s and 1s in a histogram.

sns.displot (data_bernoulli, bins = 100, kde=False)
```

## 4. Binomial Distribution

```python
# Generates 1,000 random numbers from a binomial distribution with n=100 trials and success probability p=0.2.
data_binom = stats.binom.rvs (size = 1000, n = 100, p = 0.2)

#Prints the counts of successes for each random draw.
print (data_binom)
```

```python
sns.displot (data_binom, bins = 50, kde = False)
```

## 5. Multinomial Distribution

```python
import numpy as np
```

```python
p = 1 / 6
data_multinomial = np.random.multinomial (n = 6, pvals = [p, p, p, p, p, p], size = 1000)
print (data_multinomial)
```

```python
icu = 0.3
surgery = 0.2
leave = 0.5
data_multinomial = np.random.multinomial (n = 1000, pvals = [icu, surgery, leave], size = 365)
print (data_multinomial)
```
